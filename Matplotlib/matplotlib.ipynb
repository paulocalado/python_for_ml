```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
```

```python
y = x**2
```

```python
x
```

```python
y
```

```python
#FUNCTIONAL METHOD
```

```python
plt.plot(x,y,'y-')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(x,y/3,'y')
```

```python
#OBJECT ORIENTED
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)

axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('title')
```

```python
fig = plt.figure()

axes_1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes_2 = fig.add_axes([0.2,0.6,0.4,0.3])

axes_1.set_title('Teste')
```

```python
fig,axes = plt.subplots(nrows=3,ncols=3)

plt.tight_layout()
#axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')

plt.tight_layout()
```

```python
fig = plt.figure(figsize=(8,2))

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,5))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig('myFirstPlot.png')
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

axes.plot(x,x**2, label='X Squared')
axes.plot(x,x**3, label='X Cubed')

axes.legend(loc='center left')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='#FF8C00',lw=3,alpha=0.5,ls='--',marker='o',markersize=10,markerfacecolor='blue')

ax.set_xlim([0,10])
ax.set_ylim([0,20])
```
