```python
import seaborn as sns
import matplotlib as plt
%matplotlib inline

iris = sns.load_dataset('iris')
iris.head()
```

```python
iris[iris['sepal_length']<5]['species'].unique()
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
#g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```
